```python
#Este el mi rpograma para proyecto final de la clase de algoritmos y programación
#En este proyecto buscaremos conocer la huella de carbono generada por una persona

print("--------------Bienvenido al prográma para calcular tu huella de carbono--------------")


#Para poder hallar la huella de carbono es necesario multiplicar el dato de consumo (actividad)
# por su correspondiente factor de emisión en función de el Dioxido de carbono equivalente


print("Para poder hallar efectivamente tu huella de carbono es necesario responder algunas preguntas de tu vida cotidiana")
print("------------------------------------------------------------------------------------------------------------------")

#Para hallar la huella de carbono por persona es necesario encontrar los factores de emision por cada actividad realizada los cuales son:

factores_emision = {
    "electricidad": 0.5,  # kg CO2e por kWh
    "gasolina_auto": 2.3, # kg CO2e por litro
    "gasolina_moto": 2.1, # kg CO2e por litro
    "gas": 2.0,           # kg CO2e por m3
    "carne_res": 7.2,     # kg CO2e por kg
    "carne_pollo": 4.8,   # kg CO2e por kg
    "agua": 0.09,         # kg CO2e por m3
    "viaje_auto": 0.23,   # kg CO2e por km
    "viaje_trp" : 0.2,    # Kg CO2 por km
    "viaje_moto": 0.21    # kg CO2 por km
}

huella_total = 0.0

# Acá comienzan las preguntas personales realizadas al usuario

respuesta1 = input("¿Tienes automóvil propio? (SI/NO): ").upper()

if respuesta1 == "SI":
    viaje_en_auto = float(input("¿Cuántos kilómetros al día recorres?: "))
    huella_auto = viaje_en_auto * factores_emision["viaje_auto"] * 30
    consumo_gasolina_auto = float(input("¿Cuántos litros de gasolina consumes semanalmente?: "))
    huella_gasolina_carro = consumo_gasolina_auto * factores_emision["gasolina_auto"]* 4

    huella_total += huella_auto + huella_gasolina_carro



else:
    respuesta2 = input("¿Usas moto? (SI/NO): ").upper()

    if respuesta2 == "SI":
        viaje_en_moto = float(input("¿Cuál es el recorrido en kilómetros al dia recorres?: "))
        huella_moto = viaje_en_moto * factores_emision["viaje_moto"] * 30
        consumo_gasolina_moto = float(input("¿Cuántos litros de gasolina consumes semanalmente?: "))
        huella_gasolina_moto = consumo_gasolina_moto * factores_emision["gasolina_moto"]* 4

        huella_total += huella_moto + huella_gasolina_moto



    else:
        respuesta3 = input("¿Utilizas transporte público? (SI/NO)?: ").upper()

        if respuesta3 == "SI":
            viaje_en_transp = float(input("¿Cual es el recorrido en kilometros que haces diario en el transporte público?: "))
            huella_trp = viaje_en_transp* factores_emision["viaje_trp"] * 24

            huella_total += huella_trp



consumo_gas = float(input("¿Cuantos m3 de gas consumidos aparecen en tu factura mensual ?: "))
huella_gas = consumo_gas * factores_emision["gas"]

huella_total += huella_gas



consumo_electricidad = float(input("¿Cuántos kWh de electricidad mensuales aparecen que consumes en tu recibo de la luz?: "))
huella_electricidad = consumo_electricidad * factores_emision["electricidad"]

huella_total += huella_electricidad

respuesta4 = input("¿Que tipo de carne comes mas frecuentemente de res o de pollo? (Señala tu respuesta): ").lower()
if respuesta4 == "res":
  consumo_carne_res = float(input("¿Cuántos kilogramos de carne de res consumes semanalmente?:  "))
  huella_carne_res = consumo_carne_res * factores_emision["carne_res"] * 4

  huella_total += huella_carne_res



else:
  consumo_carne_pollo = float(input("¿Cuántos kilogramos de carne de pollo consumes semanamalmente?:  "))
  huella_carne_pollo = consumo_carne_pollo * factores_emision["carne_pollo"] * 4

  huella_total += huella_carne_pollo



consumo_agua = float(input("¿Cuantos m3 de agua consumidos aparece en tu factura mensual?:  " ))
huella_agua = consumo_agua * factores_emision["agua"]

huella_total += huella_agua



#Para calcular la huella de carbono total de una persona se suman todas las huellas encontradas anteriormente


print("--------------Esas fueron todas las preguntas muchas gracias por responder----------------")
print("------------------------------------------------------------------------------------------")


print("tu huella de carbono es de", round(huella_total, 2) , "kg CO2eq/ mes")









```
